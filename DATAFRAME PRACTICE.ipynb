# DATA FRAME 

```python
raw_data = {'Cust_id' : [1,2,3,4,5],
           'First_Name': ['Alex', 'john', 'Haden','Hadelin', 'kelly'],
           'Last_Name': ['jender', 'paddison', 'hopelaman', 'robbin','jhon']}
```

```python
raw_data
```

```python
import pandas as pd
```

```python
# this command pd.DataFrame will convert any dict or data into table format and type will be dataframe.
customer_data = pd.DataFrame(raw_data)
```

```python
customer_data
```

```python
type(customer_data)
```

```python
raw_data1 = {'cust_id' : [1,2,3,4,6],
            'segment' : ['python', 'hadoop', 'scala', 'r', 'tableau'],
            'sales' : [250, 300, 275, 300, 300]}
```

```python
raw_data1
```

```python
sales_data = pd.DataFrame(raw_data1)
```

```python
sales_data
```

```python
type(sales_data)
```

```python
# JOINT TWO TABLES
# JOINT OF TWO TABLE WILL DONE BASED ON ONE COMMON COLUMN.
# JI=OINT TYPES - INNER JOIN , LEFT JOIN , RIGHT JOIN , OUTER JOIN OR FULL JOIN
# MERGE FUNCTION WILL USE FOR JOINING TWO TABLE
```

```python
inner_join = pd.merge(customer_data, sales_data, left_on = 'Cust_id', right_on = 'cust_id', how = 'inner')
```

```python
inner_join
```

```python
left_join = pd.merge(customer_data, sales_data, left_on ='Cust_id', right_on = 'cust_id', how = 'left')
```

```python
left_join #can see entire left table and common data from right table will join.
```

```python
right_join = pd.merge(customer_data, sales_data, left_on = 'Cust_id', right_on = 'cust_id', how = 'right')
```

```python
right_join # in which entire right table data and common data from left table will join, unmatch data print as NaN.
```

```python
outer_join = pd.merge(customer_data, sales_data, left_on = 'Cust_id', right_on = 'cust_id', how = 'outer')
```

```python
outer_join # in this join entire data will join from both table.
```

# converting array into dataframe

```python
import numpy as np
```

```python
np.array([['a',5,6],['b',4,7],['c',2,4]])
```

```python
df1 = pd.DataFrame(np.array([['A',5,7],['B',43,5],['C',23,4]]), columns=['Name','Attr1','Attr2'])
```

```python
df1
```

```python
df2 = pd.DataFrame(np.array([['A',5,19],['B', 4, 16],['C',2,9]]), columns = ['Name', 'Attr3','Attr4'])
```

```python
df3 = pd.DataFrame(np.array([['A',5,19],['B', 4, 16],['E', 2, 9]]), columns = ['Name', 'Attr5', 'Attr6'])
```

# Joining 3 DataFrame 

```python
pd.merge(df1,df2, on ='Name') # this will do joint with 2 table, as column name is common so left on , right on seperate write not required
```

```python
pd.merge(pd.merge(df1, df2, on = 'Name'), df3, on = 'Name') # so for fetching 3 table data 2 time use merge function
```

```python
# another way of merging 3 table

df1.merge(df2, on = 'Name').merge(df3, on = 'Name')
```

```python
# joining more than one column or multiple column from 2 table
```

```python
pd.merge(df1, df2, how = 'inner', left_on = ['Name', 'Attr1'], right_on = ['Name', 'Attr3'])
```
